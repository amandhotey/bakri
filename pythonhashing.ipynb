```python
import hashlib
given_str = b"Hello world"
given_str_2 = b"Hello world !!"

md5_value = hashlib.md5(given_str)
md5_value_2 = hashlib.md5(given_str_2)


print(md5_value.hexdigest())
print(md5_value_2.hexdigest())
```
